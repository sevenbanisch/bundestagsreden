<p style="text-align: justify"><b>1. @st.cache_data</b> = Streamlit- Dekorator für eine schnellere und effizientere Datenverarbeitung bei großer Datenmenge (Funktion wird 'gecached': wenn gleiche Parameter noch einmal abgefragt werden, müssen sie nicht neu berechnet werden)<br><br><b>2. alleReden = []</b> = leere Liste, in die später die geladenen Reden eingefügt werden können<br><br><b>3. with jsonlines.open(f'data/speeches_{legislatur}.jsonl') as f:</b> = öffnet für jede unten genannte Legislatur eine JSONL- Datei im Ordner 'data'<br><br><b>4. line["legislatur"]</b> = fügt bei geladenem Datensatz 'Legislatur' hinzu, damit man weiß aus welchem der beiden Perioden die Rede stammt<br><br><b>5. alleReden.sort(key = lambda x:x['date'])</b> = sortiert alle Reden nach Datum (notwendig, da die Reden später teilweise chronologisch analysiert werden sollen)<br><br><b>6. alleReden = load_data([19, 20])</b> = alle Reden von beiden Legislaturperioden werden ausgegeben</p>

```python
#@st.cache_data
def load_data(legislaturen):
    alleReden = []
    for legislatur in legislaturen:
        with jsonlines.open(f'../../data/speeches_{legislatur}.jsonl') as f:
            for line in f.iter():
                line["legislatur"] = legislatur
                alleReden.append(line)
    alleReden.sort(key = lambda x:x['date'])
    return alleReden

alleReden = load_data([19, 20])
```

### 3. Schritt: Reden für die Analysen vorbereiten

<p style="text-align: justify">Der Datensatz muss für die Analysen vorbereitet werden. Die Reden werden dafür verarbeitet, in einzelne Sätze zerlegt und nach bestimmten Keywords durchsucht. Dieser Schritt bildet die Grundlage für Textanalysen, Keywordsuche in den Reden und kontextbasierte Visualisierung. Durch die Definition an dieser Stelle des Codes kann später an verschiedenen Stellen auf die Funtion zurückgegriffen werden, ohne die Funtion mehrfach zu definieren.</p>

<p style="text-align: justify"><u>Genauere Erklärungen für einzelne Zeilen:</u></p>

<p style="text-align: justify"><b>1. split_into_sentences</b> = Funktion teilt Reden mit Hilfe von Satzzeichen in einzelne Sätze ein und liefert so eine Liste von Sätzen<br><br><b>2. highlight_keywords</b> = expliziete Keywords werden im Text markiert<br>(Keywords werden der Länge nach sortiert = längere Worte zuerst, sodass Doppelungen vermieden werden / <b>\b</b> = Keyword steht am Wortanfang / <b>[a-zA-ZäöüÄÖÜß]*</b> = erlaubt beliebige Wortendungen / <b>re.IGNORECASE</b> = Groß- und Kleinschreibung wird ignoriert / Keyword wird mit dem <b>mark Tag</b> farblich markiert)<br><br><b>3. extract_keyword_context</b> = sucht Sätze mit Keywords + 1 Satz davor + 1 Satz danach raus und bildet so einen Kontext für das Keyword<br>(Text wird in einzelne Sätze unterteilt / <b>keywords_lower</b> = Keywords werden in lowercase umgewandelt für case-insensitve Suche / Sätze werden nach Keyword durchsucht / falls Keyword enthalten ist, wird Kontext gebildet mit einem Satz davor und einem danach (falls vorhanden) / Kontext wird in Liste <b>contexts</b> gespeichert / Liste wird zurückgegeben)</p>

```python
#Reden in einzelne Sätze splitten
def split_into_sentences(text):
    sentences = re.split(r'(?<=[.!?]) +', text)
    return sentences

#Keywords in Sätzen markieren
def highlight_keywords(text, keywords):
    for kw in sorted(keywords, key=len, reverse=True):
        pattern = re.compile(rf'\b({re.escape(kw)}[a-zA-ZäöüÄÖÜß]*)', flags=re.IGNORECASE)
        text = pattern.sub(r'<mark>\1</mark>', text)
    return text

#Keywords + 1 Satz davor + 1 Satz danach
def extract_keyword_context(text, keywords):
    sentences = split_into_sentences(text)
    contexts = []
    keywords_lower = [kw.lower() for kw in keywords]
    for i, sentence in enumerate(sentences):
        sentence_lower = sentence.lower()
        if any(kw in sentence_lower for kw in keywords_lower):
            prev_sent = sentences[i-1] if i > 0 else ""
            next_sent = sentences[i+1] if i < len(sentences)-1 else ""
            context = "\n".join([prev_sent, sentence, next_sent]).strip()
            contexts.append(context)
    return contexts
```

<p style="text-align: justify">Die vorliegende Steamlit- Applikation erlaubt es den Nutzer*innen einen Einblick in die Umweltdebatte im deutschen Bundestag zu gewinnen.</p>

<p style="text-align: justify">Die erste Analyse zeigt, dass alle Parteien sich mit dem Thema Umwelt beschäftigen. Bedacht werden muss dabei aber, dass die häufigsten Sprecher im Bundestag in der Regel Abgeordnete der Regierungsparteien sind. Da sie mehr Mandate haben und damit auch mehr Redezeit erhalten, haben diese Parteien auch öfter die Möglichkeit Themen der Umweltdebatte anzubringen. Trotzdem sind Unterschiede zu erkennen und zwar nicht nur in der Häufigkeit, sondern auch in der Schwerpunktsetzung und Wortverwendung im Themenbereich Umweltdebatte. So sprechen beispielsweise 'DIE GRÜNEN/ BÜNDNIS 90' in 601 Reden von der 'Klimakrise' während alle anderen Parteien diesen Begriff deutlich weniger nutzen. Erwartungsgemäß benutzten die eher konservativen oder rechten Parteien ganz andere Begriffe im Kontext der Umweltdebatte. So fällt zum Beispiel der Begriff 'Klimaterroristen' insgesamt zehn Mal in den beiden vorliegenden Legislaturperioden - neun Fälle gehen dabei auf das Konto der 'AfD'.</p>

<p style="text-align: justify">Auf der Ebene der Politiker*innen gibt es Abgeordnete, die besonders häufig die Themen der Umweltdebatte ansprechen. Auch hier gibt es eine logische Erklärung für die Verteilung: Politiker*innen, die ihr Fachgebiet in der Umwelt haben, werden diese Themen deutlich öfter in ihren Beiträgen thematisieren. Durch die Kontextanalyse kann außerdem ein guter Einblick gewonnen werden, wie einzelne Politiker*innen die Keywords auf unterschiedliche Art und Weise framen, um ihren eigenen Standpunkt zu vertreten.</p>
