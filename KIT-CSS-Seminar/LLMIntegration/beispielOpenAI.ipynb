```python
import os
import openai
# Diese beiden Codes sind individuell und müssen selbst von der Website von openai besorgt werden. Leider kostet die Anwendung dieses Notebooks ein wenig Geld. Aber wirklich nicht viel!
ApiKey = os.environ['OPENAI_API_KEY']
client = openai.OpenAI(api_key=ApiKey)  # dein API-Key
```

```python
response = client.chat.completions.create(
    model="gpt-4",  # oder "gpt-3.5-turbo"
    messages=[
        {"role": "user", "content": "Was ist spaCy?"}
    ]
)

print(response.choices[0].message.content)
```

```python
import jsonlines

legislatur = 20

alleReden = []
with jsonlines.open(f'../../data/speeches_{legislatur}.jsonl') as f:
    for line in f.iter():
        #for line in list(f):
        alleReden.append(line)

alleReden.sort(key = lambda x:x['date'])
print(len(alleReden))
```

```python
welche = 24002
eineRede = alleReden[welche]
print(eineRede['text'])
```

```python
system_msg = 'Du bist eine Experte in deutscher Politik. Dein Ziel ist es, den Menschen in Deutschland aktuelle politische Diskussionen verständlicher zu machen.'

system_msg = 'Du bist ein hilfsbereiter Assistent, der Reden prägnant zusammenfassen kann.'

user_msg = f"""Kannst mir den Inhalt folgender Bundestagsrede zusammenfassen? Hier ist die Rede: '{eineRede['text']}' """

response = client.chat.completions.create(
    model="gpt-4",  # oder "gpt-3.5-turbo"
    messages=[
        {"role": "system", "content": system_msg},
        {"role": "user", "content": user_msg}
    ]
)

print(response.choices[0].message.content)
```
