```python
import os
import json
from datetime import datetime
from lxml import etree
```

```python
def loadSpeeches():
    path = '../../data/speeches_20.jsonl'
    with open(path, 'r', encoding='utf8') as fp:
        return [json.loads(line) for line in fp]

def printSpeechMetaData(speech):
    print(speech.keys())
    print(speech['id'])
    print(speech['period'])
    print(speech['date'])
    print(speech['name'])
    print(speech['party'])
    print(speech['redner_id'])
    print(speech['discussion_title'])
    
def groupSpeechesByDiscussionTitle(speeches):
    groupedSpeeches = {}
    for speech in speeches:
        tpo = speech['discussion_title']
        if tpo in groupedSpeeches:
            groupedSpeeches[tpo].append(speech)
        else:
            groupedSpeeches[tpo] = [speech]
    
    return groupedSpeeches
```

```python
speeches = loadSpeeches()
first = speeches[0]
print(len(speeches))

grouped = groupSpeechesByDiscussionTitle(speeches)
tpos = list(grouped.keys())
print(tpos[0])
print(len(grouped[tpos[0]]))
print(*grouped[tpos[0]], sep='\n')
```
